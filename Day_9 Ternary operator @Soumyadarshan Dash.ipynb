### Ternary operator

Python's ternary operator is a really helpful way of saying "if this, then that" in just one line of code. It's great for keeping code tidy, especially when you need to check lots of different things.

**For example**, if a variable called 'grade' is assigned a value between 0 and 100, we can use the ternary operator to assign 'Pass' to grade if it is greater than or equal to 50, and 'Fail' if it is less than 50, as shown below:

```python
grade = int(input("Enter student grade"))
```

**if and else Method**

```python
if grade >=50:
    print("Pass")
    
else:
    print ("Fail")
```

**Ternary operator**

```python
grade = 'Pass' if grade >= 50 else 'Fail'
```

```python
print(grade)
```

**Note:**

- The main difference between the if/else statement and the ternary operator is that the ternary operator is a more concise syntax for writing if/else statements. The ternary operator can be used to simplify an if/else statement, while still maintaining the same functionality. Additionally, the ternary operator is only valid for expressions that have exactly one true or false value, whereas the if/else statement can handle multiple conditions.

Think you are the manager of a store and you have to give a discount to customers based on their purchase amount.

- if a customer spends more than $100, you want to give them a 10% discount.

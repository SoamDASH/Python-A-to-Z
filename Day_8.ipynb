#### if…else statement

The if-else statement in Python allows you to test a condition and execute different code blocks based on the outcome.

- The 'if-else' statement in Python is like a choose our own adventure story.
we give the computer a question and it gives you two choices. 


- If the anser is yes, the computer will do one thing, if the answer is no, the computer will do something else.


- The **'if'** statement is like asking the question, and the code that come after it is like what the computer will do if the answer yes.


- The **'else'** statement is like the other choice the computer has, and the code that comes after it is what the computer will do if the answer is no.


**Like**

```python
if this is true:
    do this
else:
    do that
```

Exmple

```python
number = 5
if number > 10:
    print("The number is bigger than 10")
else:
    print("The number is smaller than 10")
```

- In this example, the code block after else will be executed as the number 5 is smaller than 10.




- The if-else statement is a way for the computer to make decisions, and it is a fundamental concept in programming.

**We also discuss nested blocks, but please be patient.**

**The if-elif-else statement in Python allows you to test multiple conditions and execute different code blocks based on the outcome.**

```python
if condition1:
    # code to execute if condition1 is True
elif condition2:
    # code to execute if condition1 is False and condition2 is True
elif condition3:
    # code to execute if condition1 and condition2 are False and condition3 is True
else:
    # code to execute if all of the conditions are False
```

-  It's important to know that the code after the if statement will only be executed if the answer is yes and the code after the else statement will only be executed if the answer is no.

But The elif statement stands for "else if" and it's like asking a second question, and the code that comes after it is like what the computer will do if the first answer is no and this second answer is yes. 

**Note** the first if statement will only be executed if the answer is yes, the code after the first elif statement will only be executed if the first answer is no and this second answer is yes and so on, and the code after the else statement will only be executed if all the previous answers are no.

**For Example**

```python
temperature = 75
if temperature > 90:
    print("Wear shorts and t-shirt")
elif temperature > 60:
    print("Wear a long-sleeve shirt and pants")
else:
    print("Wear a coat and a scarf")
```

- - In this example, the code block after the first elif will be executed as the temperature is greater than 60 and less than 90.

It's a way for the computer to make multiple decisions based on the input.

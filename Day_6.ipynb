```python
import math

#input initial velocity, angle, and gravitational acceleration
velocity = float(input("Enter the initial velocity (m/s): "))
angle = float(input("Enter the angle of projection (degrees): "))
g = int(input("Enter the gravitational acceleration input value should be integer(m/s^2): "))

#convert angle to radians
angle = math.radians(angle)

#compute the time of flight
time_of_flight = (2 * velocity * math.sin(angle)) / g

#compute the horizontal distance
horizontal_distance = velocity * math.cos(angle) * time_of_flight

#compute the maximum height
maximum_height = (velocity * math.sin(angle))**2 / (2 * g)

#print the results
print("Time of flight: ", time_of_flight)
print("Horizontal distance: ", horizontal_distance)
print("Maximum height: ", maximum_height)
print("Code design by: Soumyadarshan Dash")
```

### Boolen

- In Python, a bool is a special kind of data that can only be either true or false. This is useful in computer programming to show if something is correct or not. For example, if a certain condition is met, a bool variable can be set to true, and if it is not met, it can be set to false. These true or false values are often used in computer code to make decisions and compare values using special operators like "==" or "!=".

**Example**

```python
#define some variables
x = 5
y = 10

#compare the values of x and y using a comparison operator
result = x < y
print(result) # Output: True

#use the boolean value stored in 'result' to make a decision
if result:
    print("x is less than y.")
else:
    print("x is greater than or equal to y.")
```

In this example, we have two numbers, x is 5 and y is 10. We use a special symbol called < to check if x is smaller than y. This check gives us a true or false answer that we store in a variable called result.
Next, we use a special command called if-else to look at the answer stored in result, if it's true we print one sentence, if it's false we print another one.
In this case, the answer is true, x is smaller than y, so the message printed is "x is less than y.".
Booleans are often used with these special symbols to make choices in a computer program.

```python
a = True
print(bool(a)) # Output: True
b = "hello"
print(bool(b)) # Output: True
c = ""
print(bool(c)) # Output: False
```

We can see that variable a holds a boolean value and the bool() function returns True for it, variable b holds a non-empty string, the bool() function returns True for it, variable c holds an empty string and the bool() function returns False for it

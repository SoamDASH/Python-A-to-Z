### integers


An integers is special kind of number that you can use in computer programing.Integers can be positive (like 5), negative (like -3) or Zero (like 0).

They are special because they can't have any numbers after the decimal point.

So, we can't have **integers** like 5.5. In this case 5.5 is a decimal value so we called **float**.

integers can be used to do math problems like adding,subtracting,multiplying, and dividing.


For example,

If we want to find out how many apples we have after we picked 5 from the tree and then gave away 2, we would use integers.

We would start with 5 appels and then subtract 2 so we would have 3 apples left.

So, integers are basic kind of numbers that have no fractional part, they can be positive, negatice, or zero and they are used in mathmatical operations.

So, now,let's try to understand this example in python code


Here's an example of how you could use integers to find out how many apples you have after picking 5 from a tree and then giving away 2:

```python
# Start with 5 apples
apples = 5

# Pick 5 more apples from the tree
apples += 5

# Give away 2 apples
apples -= 2

# Print the final number of apples
print(apples)
```

This code uses the += and -= operators to add and subtract from the apples variable, respectively. These operators are shorthand for apples = apples + 5 and apples = apples - 2.

The final output of this program is 8, which represents the final number of apples you have after picking 5 from the tree and giving away 2.
